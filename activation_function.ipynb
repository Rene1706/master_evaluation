```python
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def softplus(x):
    return np.log(1 + np.exp(x))

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def swish(x):
    return x * sigmoid(x)

def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x ** 3)))

# Generate a range of input values
x = np.linspace(-10, 10, 1000)

# Create a plot for each activation function
plt.figure(figsize=(12, 8))

# Sigmoid
plt.subplot(2, 4, 1)
plt.plot(x, sigmoid(x))
plt.title("Sigmoid")
plt.grid(True)

# Tanh
plt.subplot(2, 4, 2)
plt.plot(x, tanh(x))
plt.title("Tanh")
plt.grid(True)

# ReLU
plt.subplot(2, 4, 3)
plt.plot(x, relu(x))
plt.title("ReLU")
plt.grid(True)

# Leaky ReLU
plt.subplot(2, 4, 4)
plt.plot(x, leaky_relu(x))
plt.title("Leaky ReLU")
plt.grid(True)

# Softplus
plt.subplot(2, 4, 5)
plt.plot(x, softplus(x))
plt.title("Softplus")
plt.grid(True)

# ELU
plt.subplot(2, 4, 6)
plt.plot(x, elu(x))
plt.title("ELU")
plt.grid(True)

# Swish
plt.subplot(2, 4, 7)
plt.plot(x, swish(x))
plt.title("Swish")
plt.grid(True)

# GELU
plt.subplot(2, 4, 8)
plt.plot(x, gelu(x))
plt.title("GELU")
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams

# Enable LaTeX rendering
rcParams['text.usetex'] = True

# Define the activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def softplus(x):
    return np.log(1 + np.exp(x))

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def swish(x):
    return x * sigmoid(x)

def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x ** 3)))

# Generate a range of input values
x = np.linspace(-10, 10, 1000)

# Create a plot for each activation function
fig, axs = plt.subplots(2, 4, figsize=(14, 10))

# List of activation functions and their LaTeX formulas
activation_functions = [
    (sigmoid, r"$\sigma(x) = \frac{1}{1 + e^{-x}}$", "Sigmoid"),
    (tanh, r"$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$", "Tanh"),
    (relu, r"$f(x) = \max(0, x)$", "ReLU"),
    (leaky_relu, r"$f(x) = \alpha x, \, x < 0; \, f(x) = x, \, x \geq 0$", "Leaky ReLU"),
    (softplus, r"$f(x) = \log(1 + e^x)$", "Softplus"),
    (elu, r"$f(x) = \alpha(e^x - 1), \, x < 0; \, f(x) = x, \, x \geq 0$", "ELU"),
    (swish, r"$f(x) = x \cdot \sigma(x)$", "Swish"),
    (gelu, r"$f(x) = 0.5 x (1 + \tanh[\sqrt{\frac{2}{\pi}}(x + 0.044715x^3)])$", "GELU")
]

# Plot each activation function
for i, (func, latex_formula, name) in enumerate(activation_functions):
    row, col = divmod(i, 4)
    ax = axs[row, col]
    
    # Plot the activation function
    ax.plot(x, func(x))
    ax.set_title(name)
    ax.grid(True)
    
    # Add axis labels
    ax.set_xlabel('Input (x)', fontsize=10)
    ax.set_ylabel('Output f(x)', fontsize=10)

    # Add LaTeX formula as text inside the plot
    ax.text(0.05, 0.9, latex_formula, fontsize=12, ha='left', va='top', transform=ax.transAxes, usetex=True)

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```
