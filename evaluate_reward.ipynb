```python
import matplotlib.pyplot as plt

# Define the ExponentialMovingAverage class
class ExponentialMovingAverage:
    def __init__(self, alpha=0.9):
        self.alpha = alpha
        self.value = None

    def update(self, new_value):
        if self.value is None:
            self.value = new_value
        else:
            self.value = self.alpha * self.value + (1 - self.alpha) * new_value
        return self.value
```

```python
# Instantiate the ExponentialMovingAverage object
ema_psnr = ExponentialMovingAverage(alpha=0.8)

# Generate some test data (with some spikes)
data = [0, 10, 12, 15, 18, 14, 13, 0.2, 16, 19, 20, 21, 25, 28, 30, 4, 25, 22, 20]

# Apply the EMA to the data
smoothed_values = []
for value in data:
    smoothed_value = ema_psnr.update(value)
    smoothed_values.append(smoothed_value)

# Plot the original data and the smoothed values
plt.plot(data, label="Original Data", marker='o')
plt.plot(smoothed_values, label="Smoothed Data (EMA)", marker='x')
plt.title("Exponential Moving Average Smoothing")
plt.xlabel("Iteration")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```
